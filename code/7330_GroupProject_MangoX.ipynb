## <center>Period Poverty: The problem troubling women around the world</center>

<font face="Times New Roman" size=3>Up to 500 million women and girls in the world experience 'period poverty' almost every month. Due to financial constraints, they face the inability to obtain basic tools for menstrual hygiene management (eg: sanitary products and washing facilities). In many communities, the shame of menstruation often aggravates menstrual poverty and worsens chronic poverty. It also prevents these women from having the best hygiene habits. If women cannot be provided with adequate long-term support, the health and safety of women and girls Will be threatened, so assisting women with decent menstrual management is an urgent and necessary action.</font>

# Code and Visualization

```python
%matplotlib inline

#pip install dexplot

import pandas as pd 
import numpy as np
import dexplot as dxp

noworkfuringperiod = pd.read_csv('../data/' + 'Proportion_of_women_and_adolescent_girls__15-49__who_did_not_participate_in_school__work_or_social_activities_during_their_last_menstrual_period.csv')
haveprivateplacetochange = pd.read_csv('../data/' + 'Proportion_of_women_and_adolescent_girls__15-49__with_a_private_place_to_wash_and_change_and_using_menstrual_hygiene_materials.csv')

print(type(noworkfuringperiod))
noworkfuringperiod.head()
```

```python
#merge the two tables
table_2t=pd.merge(noworkfuringperiod, haveprivateplacetochange,on=["REF_AREA_DESC","TIME_PERIOD","SEX_DESC","AGE_DESC","URBANIZATION_DESC","INCOME_WEALTH_QUANTILE_DESC","TIME_DETAIL"])
table_2t.head()
```

```python
table_2t.info
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

table_total = table_2t[(table_2t['AGE_DESC'] == '15 to 49 years old') & (table_2t['URBANIZATION_DESC']=='Total') & (table_2t['INCOME_WEALTH_QUANTILE_DESC']=='Total (national average) or no breakdown')]
table_total.head()
```

```python
table_area = pd.pivot_table(table_total, values='DIDN’T_PARTICIPATE_ACTIVITIES_DURING_PERIOD_VALUE', index=['REF_AREA_DESC'])
plt.style.use('ggplot')

ax = table_area.plot.barh(width=0.5, fontsize='16',figsize=(11,6),legend=False,title='Proportion of females 15-49 who did not participate activities during last period');

for p in ax.patches:
    w = p.get_width()
    ax.annotate(f'{w:.2f}', (w * 0.5, p.get_y() + 0.1),color="w")
```

```python
table_area2 = pd.pivot_table(table_total, values='WITH_A_PLACE TO_USE_MENSTRUAL_HYGIENE_MATERIAL_VALUE', index=['REF_AREA_DESC'])

ax2 = table_area2.plot.barh(width=0.5, color='g',fontsize='16',figsize=(11,6),legend=False,title='Proportion of females with a place to use menstrual hygiene material')

for p in ax2.patches:
    w = p.get_width()
    ax2.annotate(f'{w:.2f}', (w * 0.5, p.get_y() + 0.1),color="w")
```

```python
dxp.kde(x='DIDN’T_PARTICIPATE_ACTIVITIES_DURING_PERIOD_VALUE', y='WITH_A_PLACE TO_USE_MENSTRUAL_HYGIENE_MATERIAL_VALUE', data=table_total,
       title='Did not participate activities vs. Have a place to use menstural material')
```

```python
table_age=table_2t[table_2t['AGE_DESC'] != '15 to 49 years old']
table_age.head()
```

```python
table_age2 = pd.pivot_table(table_age, values='DIDN’T_PARTICIPATE_ACTIVITIES_DURING_PERIOD_VALUE',index=['REF_AREA_DESC'], columns=['AGE_DESC'], aggfunc=np.mean)
table_age_df = pd.DataFrame(table_age2)

ax = table_age_df.plot.barh(width=0.8, fontsize='16',  alpha=0.7, figsize=(20,25),cmap='viridis')
ax.legend(bbox_to_anchor=(1, 0.5), fontsize='12');
plt.style.use('ggplot')
plt.legend(fontsize='15')
```

```python
table_urbanization = table_2t[(table_2t['AGE_DESC'] == '15 to 49 years old') 
                              & (table_2t['URBANIZATION_DESC']!='Total') 
                              & (table_2t['INCOME_WEALTH_QUANTILE_DESC']=='Total (national average) or no breakdown')]
table_urbanization.head()
```

```python
table_urbanization_use = table_urbanization.pivot_table(values = 'WITH_A_PLACE TO_USE_MENSTRUAL_HYGIENE_MATERIAL_VALUE', columns = 'URBANIZATION_DESC', index = 'REF_AREA_DESC')
table_urban_use = pd.DataFrame(table_urbanization_use)
table_urban_use.plot.barh(title = 'Proportion of females with a place to use menstrual hygiene material',figsize=(11,6),alpha=0.5)
plt.legend(fontsize='10',bbox_to_anchor=(1, 0.2),loc='best')
```

```python
table_urbanization_use.plot(kind = 'hist', figsize=(10,6),bins = 20,alpha=0.5, title = 'Proportion of females with a place to use menstrual hygiene material')
plt.style.use('ggplot')
plt.legend(fontsize='15')
plt.show()
```

```python
table_urbanization_use.plot(kind = 'box', figsize=(10,6),title = 'Proportion of females with a place to use menstrual hygiene meterial')
plt.style.use('ggplot')
plt.show()
```

```python
table_urbanization_rest = table_urbanization.pivot_table(values = 'DIDN’T_PARTICIPATE_ACTIVITIES_DURING_PERIOD_VALUE', columns = 'URBANIZATION_DESC', index = 'REF_AREA_DESC')

table_urban_rest = pd.DataFrame(table_urbanization_rest)
table_urban_rest.plot.barh(title = 'Proportion of females who did not participate activities during last period',figsize=(10,6),alpha=0.5)
plt.legend(loc='best')
```
